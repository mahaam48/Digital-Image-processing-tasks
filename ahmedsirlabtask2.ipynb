```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load your image
img = cv2.imread('task2image.jpeg', cv2.IMREAD_GRAYSCALE)

plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')
```

```python
# Convert to binary using Otsu thresholding
_, binary = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

plt.imshow(binary, cmap='gray')
plt.title("Binary Image")
plt.axis('off')

```

```python
# Perform CCL
num_labels, labels = cv2.connectedComponents(binary)

print("Total objects detected:", num_labels - 1)  # minus background




```

```python
# Map component labels to random colors
label_hue = np.uint8(179 * labels / np.max(labels))
blank_ch = 255 * np.ones_like(label_hue)
colored_img = cv2.merge([label_hue, blank_ch, blank_ch])
colored_img = cv2.cvtColor(colored_img, cv2.COLOR_HSV2RGB)

plt.imshow(colored_img)
plt.title("Labeled Image")
plt.axis('off')

```
